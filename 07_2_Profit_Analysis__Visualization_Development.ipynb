**Modification Needed:** 'region' - Region and 'Asia Pacific' to 'Asia'

```python
# Filter the DataFrame
df_asia = df_1[df_1['Region'] == 'Asia']

# Print the first 5 rows of the filtered DataFrame
(df_asia)
```

```python
# ALL country
# Group the DataFrame by 'Country' and calculate the count, sum, and mean
df_grouped = df.groupby('Country').agg(
    Count=('Country', 'count'),
    Total_Profit=('Total Profit', 'sum'),
    Average_Profit=('Total Profit', 'mean')
)

# Print the grouped DataFrame
print(df_grouped)
```

```python
# Asia region df_asia
# Group the DataFrame by 'Country' and calculate the count, sum, and mean
df_asia_grouped = df_asia.groupby('Country').agg(
    Count=('Country', 'count'),
    Total_Profit=('Total Profit', 'sum'),
    Average_Profit=('Total Profit', 'mean')
)

# Print the grouped DataFrame
print(df_asia_grouped)
```

What `agg` means?

Let's look computing *total profit*, *average profit* and *count number of countries* separately.

```python
df_asia_grouped = df_asia.groupby('Country')['Total Profit'].sum().reset_index(name='Total Profit')
df_asia_grouped
```

```python
df_asia_grouped = df_asia_grouped.groupby('Country')['Total Profit'].mean().reset_index(name='Average Profit')
df_asia_grouped
```

```python
df_asia_grouped = df_asia.groupby('Country')['Country'].count().reset_index(name='Count')
df_asia_grouped
```

**Consolidated approach - single operation that combines all these metrics using `agg`**

```python
# Asia region df_asia
# Group the DataFrame by 'Country' and calculate the count, sum, and mean
df_asia_grouped = df_asia.groupby('Country').agg(
    Count=('Country', 'count'),
    Total_Profit=('Total Profit', 'sum'),
    Average_Profit=('Total Profit', 'mean')
)

# Print the grouped DataFrame
print(df_asia_grouped)
```

### Now, let's plot a graph for Total profit for each country in Region Asia

```python
df_asia_grouped
```

```python
import matplotlib.pyplot as plt

# Plotting the total profit by country
df_asia_grouped['Total_Profit'].plot(kind='bar', color='skyblue')

plt.title('Total Profit by Country in Asia')
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting the average profit by country
df_asia_grouped['Average_Profit'].plot(kind='bar', color='skyblue')

plt.title('Average Profit by Country in Asia')
plt.xlabel('Country')
plt.ylabel('Average Profit')
plt.xticks(rotation=45)
plt.show()
```
